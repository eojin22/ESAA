# [개념]

- **np.unique(x)** : 데이터 모음(x) 안에서 중복된 값을 모두 제거하고 고유한 값들만 추려서 반

- **groupby( )** : 데이터를 그룹으로 묶어줌.

- **pd.merge(df1, df2, on, how)**: 두 개의 데이터프레임(df1, df2)을 특정 기준으로 병합
   - on: 기준이 되는 열 이름
   - how: 병합 방식

- **sns.pointplot( )** : 평균을 점으로 찍고, 선으로 이어서 추세를 보여주는 seaborn 함수

- **df.pivot( )** : 긴 형태(long) 데이터를 넓은 형태(wide)로 재배




# 새로운 전처리/시각화

**[아침 vs 밤 시간대별 장바구니 차이 분석]**

'아침'과 '밤'이라는 특정 시간대를 정의하고, 해당 시간대에 어떤 Department의 상품들이 잘 팔리는지 분석(시간대별 인기 카테고리 파악)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. 데이터 불러오기
order_products_train_df=pd.read_csv("/content/drive/MyDrive/order_products__train.csv")
order_products_prior_df=pd.read_csv("/content/drive/MyDrive/order_products__prior.csv")
orders_df=pd.read_csv("/content/drive/MyDrive/orders.csv")
products_df=pd.read_csv("/content/drive/MyDrive/products.csv")
departments_df=pd.read_csv("/content/drive/MyDrive/departments.csv")
```

```python
# 2. 데이터 합치기
# 1) 주문 테이블(시간 포함)과 장바구니 테이블 합치기
merged_df = pd.merge(orders_df[['order_id', 'order_hour_of_day']],
                     order_products_prior_df[['order_id', 'product_id']],
                     on='order_id', how='inner')

# 2) 상품 테이블 합치기
merged_df = pd.merge(merged_df, products_df[['product_id', 'department_id']], on='product_id', how='inner')

# 3) 부서(카테고리) 테이블 합치기
merged_df = pd.merge(merged_df, departments_df, on='department_id', how='inner')

merged_df.head()
```

```python
# 3. 시간대별로 그룹을 나누기(Morning vs Night)
# =====================================================================
print("Categorizing by order_hour_of_day...")

# Morning: 06:00 ~ 09:00 / Night: 22:00 ~ 02:00
conditions = [
    (merged_df['order_hour_of_day'] >= 6) & (merged_df['order_hour_of_day'] <= 9),
    (merged_df['order_hour_of_day'] >= 22) | (merged_df['order_hour_of_day'] <= 2)
]
choices = ['Morning (06-09h)', 'Night (22-02h)']

# Create a new column 'time_group'
merged_df['time_group'] = np.select(conditions, choices, default='Other')

# Filter out 'Other' times to focus only on Morning and Night
target_df = merged_df[merged_df['time_group'] != 'Other']


target_df.head()
```

```python
# 4. 시각화 (그래프 그리기)

# 가장 많이 팔린 상위 12개 카테고리만 추출
top_departments = target_df['department'].value_counts().head(12).index
plot_df = target_df[target_df['department'].isin(top_departments)]

plt.figure(figsize=(14, 8))

# countplot
sns.countplot(data=plot_df, y='department', hue='time_group',
              palette={'Morning (06-09h)': '#FFB000', 'Night (22-02h)': '#332288'},
              order=top_departments)

plt.title('Morning vs. Night Owls: Popular Departments Comparison', fontsize=18, fontweight='bold')
plt.xlabel('Total Number of Products Sold', fontsize=12)
plt.ylabel('Department', fontsize=12)
plt.legend(title='Order Time', loc='lower right', fontsize=11)

plt.show()
```
